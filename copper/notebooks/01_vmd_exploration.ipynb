# Signal Processing: Variational Mode Decomposition (VMD)
**Objective:** Raw commodity prices are highly chaotic, driven by high-frequency algorithmic day trading. Before feeding data into our Deep Learning engine, we must separate this daily "noise" from the true underlying macroeconomic "signal" (the slow-moving physical supply chain trend).

We utilize **VMD** to decompose the Dr. Copper price proxy into $K$ discrete Intrinsic Mode Functions (IMFs).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from vmdpy import VMD

# 1. Load the raw data
df = pd.read_parquet('../data/processed/master_feature_set.parquet')
copper_signal = df['copper_close'].values

# 2. Set VMD Mathematical Parameters
alpha = 2000       # moderate bandwidth constraint
tau = 0.           # noise-tolerance (no strict fidelity enforcement)
K = 4              # number of modes to extract
DC = 0             # no DC part imposed
init = 1           # initialize omegas uniformly
tol = 1e-7         # tolerance of convergence

# 3. Run the Decomposition
u, u_hat, omega = VMD(copper_signal, alpha, tau, K, DC, init, tol)

# 4. Visualize the Extraction
plt.figure(figsize=(15, 10))
plt.subplot(K+1, 1, 1)
plt.plot(copper_signal, color='black', label='Original Copper Price (Raw & Chaotic)')
plt.legend(loc='upper left')
plt.title('Variational Mode Decomposition of Dr. Copper')

colors = ['red', 'orange', 'blue', 'green']
titles = ['IMF 1: The Macro Trend', 'IMF 2: Medium Cycle', 'IMF 3: Fast Cycle', 'IMF 4: High-Frequency Noise']

for i in range(K):
    plt.subplot(K+1, 1, i+2)
    plt.plot(u[i,:], color=colors[i], label=titles[i])
    plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
